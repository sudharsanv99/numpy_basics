```python
# statistics
# If axis = 0, then it  means that each column
x = np.array([[1,2,4,5,6], [1,2,3,4,5]])
y = np.array([[2,34,5,6,3],[1,2,3,4,5]])
np.max(x, axis = 0)
```

```python
# If axis = 1, then it means each row
np.max(x, axis = 1)
```

```python
np.min(y,axis = 1)
```

```python
# Sum:
print(np.sum(x))
print(np.sum(x, axis= 1)) # Sum of each row
print(np.sum(x, axis= 0)) # Sum of each column
```
