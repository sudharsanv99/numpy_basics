```python
import numpy as np
```

```python
file_data = np.genfromtxt("data.txt", delimiter=",")
print(file_data)
```

### Boolean Masking and Advanced Indexing

```python
# Find the filedata which has values greater than 50
print(file_data)
```

```python
# This shows all the values which has values greater than 50
file_data[file_data>50]
```

```python
#Tells if any column has values greater than 50
np.any(file_data>50,  axis = 0)
```

```python
# If all column / row has values greater than 50
np.all( file_data>30, axis  = 1)
```

```python
# Multiple conditions:
(file_data > 50) & (file_data < 100)
```

```python
# Negation:
print(~((file_data > 50) & (file_data < 100)))
```
