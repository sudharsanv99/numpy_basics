```python
import numpy as np
```

```python
a = np.array([1.0, 2.0, 3.0], dtype = "int16")
a.dtype
```

```python
# 2D Arrays
b = np.array([
    [1,2,3,4,5], 
    [2,3,4,5,6], 
    [5,6,78,8,9], 
    [21,44,212,34,45]])
b.shape
```

```python
# Get a specific element -> 212
print(b[3][2])
```

```python
# Get a specific row:
print(b[1,:])

# Get a specific column:
print(b[:,2])
```

```python
# Get a sub array using slicing:
# Prints the sliced sub array in the format - startIndex:endIndex:stepValue
print(b[1, 0:-1:2])
```

```python
# Change a value of the array:
b[1][2] = 10
b
```

```python
# Changes the entire row
b[1,:] = 3
print(b)
```

```python
# Replace in a 2D array:
b[:,2] = [1,23,33,12]
b
```

```python
# Access specific rows and columns
x = np.array([[1,2,3,4,5],[6,7,4,89,3],[12,12,13,45,67],[45,6,64,32,56]])
print(x)

print(x[[0,1,2],[1,2,3]])

print(x[[0,2,3],3:])
```

```python
#3D Arrays:
c = np.array([
    [
        [1,2,3],
        [4,5,6]
    ],
    [
        [10,11,12],
        [11.1,12,13]
    ]
]) 
print(c)
```

```python
print(c.ndim)
print(c.dtype)
print(c.itemsize)
print(c.size)
print(c.nbytes)
print(c.shape)
```

```python
# Change data in a 3D array: (Work outside in)
c[1,1,0] = 11
print(c)
```

```python
# Replace in 3D:
#But if the replacement value does not have the same dimensions  - then throws a ValueError
c[:,1,:] = [[1,2,3], [4,5,6]]
print(c)
```
