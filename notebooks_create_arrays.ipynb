Numpy - Different forms of array creation

```python
import numpy as np
```

```python
# All zeroes matrix:
np.zeros(5)
```

```python
np.zeros((2,3))
```

```python
np.zeros((2,4,3))
```

```python
# All one datatype:
np.ones((2,2,3), dtype = "int32")
```

```python
# Any other number - full
np.full((2,3), 90, dtype = "int16")
```

```python
a = np.array([[21,32,32], [21,34,23]])
```

```python
# full using existing array shape:
np.full(a.shape,40)
```

```python
# Full like:
np.full_like(a, 40)
```

```python
# Random decimal numbers:
np.random.rand(2,3)
```

```python
# Random  Sample
np.random.random_sample(a.shape)
```

```python
np.ones(a.shape)
```

```python
# Random Int values:
np.random.randint(-10, 100, size = (2,3,3))
```

```python
# Identity Matrix:
np.identity(4)
```

```python
#Repeat an array:
arr = np.array([[1,2,4]])
np.repeat(arr,3,axis  = 0)
```

```python
a =  np.zeros((5,5), dtype = "int32")
a[0,:] = 1
a[:,0] = 1
a[4,:] = 1
a[:,4] = 1
a[2,2] = 9
print(a)
```

```python
b = np.ones((5,5), dtype= "int32")
c = np.zeros((3,3))
c[1,1] = 9
b[1:-1, 1:-1] = c
print(b)
```
